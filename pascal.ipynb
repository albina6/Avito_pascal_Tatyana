```python
var needCh := ['б', 'в', 'г', 'д', 'ж', 'з'];
textIn : string;
chOddWord : set of char;
index, i, len: integer;

begin
  index := 1;
  write('Введите текст --> ');
  read(textIn);
  len := length(textIn);
  for i:=1 to len do
    begin
      if textIn[i] = ' '
        then index := index + 1
      else
        begin
          if index mod 2 = 0
            then needCh := needCh - [textIn[i]] 
          else
            chOddWord :=chOddWord + needCh * [textIn[i]] ;
        end;
      
    end;
  chOddWord :=chOddWord * needCh;
  chOddWord.ToString.Println
end.
```

```python
type SetString = set of string;
var
  n, cntFarm, index: integer;
  product : string;
  productSet, answer1, answer2 : SetString;
  nFarm : array of SetString;

begin
  writeln('Последовательно введите не пустой ',
      'список сельскохозяйственных культур:');
  readln(product);
  repeat
    productSet += [product];
    readln(product);
  until length(product) = 0;
  write('Введите количество колхозов --> ');
  readln(n);
  nFarm := new SetString[n];
  for var i:= 0 to n-1 do
  begin
    writeln(NewLine + 'Колхоз номер ', i+1);
    writeln('Последовательно введите не пустой список ',
       'сельскохозяйственных культур, которые выращивают в колхозе');
    readln(product);
    repeat
      nFarm[i] += [product];
      readln(product);
    until length(product) = 0;
  end;
  
  answer1 := productSet;
  for var i:= 0 to n-1 do
  begin
    answer1 *= nFarm[i]
  end;
  answer1.ToString.Println;
  
  write(NewLine + 'Введите количество колхозов, ',
      'в которых хотите увидеть общие сельскохозяйственные культуры: ');
  readln(cntFarm);
  answer2 := productSet;
  writeln('Последовательно введите номера тех колхозов, ',
      'в которых хотите увидеть общие сельскохозяйственные культуры:');   
  for var i:= 0 to cntFarm-1 do
  begin
    readln(index);
    answer2 *= nFarm[index - 1];
  end;
  writeln(NewLine + 'Общие культуры для введенных колхозов: ');
  foreach var s in answer2 do
    writeln(s);

end.
```

```python
var 
  myArray: array[1..100] of integer;
  i, ind_min, ind_max, min_el, max_el : integer;

begin
  write('Числа: ');
  for i:= 1 to 100 do 
  begin
    myArray[i] := random(-100, 100);
    write(myArray[i], ' ');
  end;
  
  ind_min := 1;
  ind_max := ind_min;
  min_el := myArray[ind_min];
  max_el := myArray[ind_min];
  
  for i:= 1 to 100 do 
  begin
    if myArray[i] < min_el then
    begin
      ind_min := i;
      min_el := myArray[i];
    end    
    else if myArray[i] > max_el then
    begin
      ind_max := i;
      max_el := myArray[i];
    end;
  end;
  writeln(NewLine + NewLine + 'Минимальный элемент в массиве расположен по индексу ',
    ind_min, ' и равен ', min_el);
    
  writeln('Максимальный элемент в массиве расположен по индексу ',
    ind_max, ' и равен ', max_el);
end.
```
